```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8          
E0 = 1.0 # E field amp.
wavelength = 1.0 
k = 2*np.pi / wavelength   # wave number
betas = np.linspace(0.01, 0.9999, 100)  
v = betas * c

# Galilean residual R = (k * v * E0) / c^2
R_gal = (k * v * E0) / (c**2)

# Lorentz residual is R =0 
R_lor = np.zeros_like(betas)

# Plot
plt.plot(betas, R_gal, label="Galilean residual")
plt.plot(betas, R_lor, label="Lorentz residual")
plt.xlabel("Beta = v/c")
plt.ylabel("Residual amplitude")
plt.title("Residual of Ampere-Maxwell law")
plt.legend()
plt.grid(True)
plt.show()
```
