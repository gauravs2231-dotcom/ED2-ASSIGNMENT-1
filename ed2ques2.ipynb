```python
import numpy as np
import matplotlib.pyplot as plt

beta = [0.01, 0.5, 0.9999] 
x = np.array([[1], [0]])     # column vector (1 0) x axis
y = np.array([[0], [1]])     #column vector (0 1)  y axis 


def plotv(v, c, sty, lab):
    plt.plot([0, v[0,0]], [0, v[1,0]], c+sty, label=lab)
    #the above line starts ploting from 0(origin) to our prescribed point on that axis ( like in my case for x axis [0,1] )

for b in beta:
    theta = np.arctan(b) 
    # Passive anticlockwise rotation matrix R
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta),  np.cos(theta)]]) 

    # rotated vectors
    x1 = R @ x
    y1 = R @ y

    # plot
    plt.figure()
    plotv(x, 'b', '-', 'X')
    plotv(y, 'g', '-', ' Y or TIME')
    plotv(x1, 'b', '--', 'Rotated X')
    plotv(y1, 'g', '--', 'Rotated Y')

    plt.xlim(-0.3, 1)
    plt.ylim(-0.5, 2)
    plt.title(f"Rotation with Î²={b}")
    plt.legend()
    plt.grid(True)
    plt.show()


```
