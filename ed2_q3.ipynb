```python
import matplotlib.pyplot as plt 
import numpy as np

b=np.linspace(0,0.99999,300)  

def g(x):
    return( 1 / (np.sqrt(1 - b**2))  )  # basically definig gamma 
    

beta = np.array([0.01, 0.5, 0.9999])
gamma = 1 / np.sqrt(1 - beta**2)
labels = ["0.01c", "0.5c", "0.9999c"]

#PLotting IMproper time / Proper time (i.e Gamma)   v/s Beta(V/c)
plt.figure() 
plt.axhline(0, color="b", linewidth=0.7) 
plt.axvline(0, color="k", linewidth=0.7) 
plt.ylim(top=30)
plt.plot(b, g(b), label="RATIO OF IMPROPER TO PROPER TIME ")
plt.xlabel(" β=V/C ")
plt.ylabel("GAMMA γ ")
plt.title("  Fig1: Time dilation: γ vs β") 
plt.legend()
plt.show()


#PLotting specific points as per question
plt.scatter(beta, gamma, color='red')
for i, label in enumerate(labels):
    plt.text(beta[i]+0.01, gamma[i], label)  # label points
plt.xlabel("βeta = v/c")
plt.ylabel("GAMMMA γ ")
plt.title("Fig 2: Time dilation: γ vs β")
plt.grid(True)
plt.ylim(top=100)
plt.legend()
plt.show()
```
