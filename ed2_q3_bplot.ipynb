```python
import numpy as np
import matplotlib.pyplot as plt

tau0 = 1.0   # Lifetime in particles own rest frame 
n0 = 1.0     # Inintial no. of particles 
beta = np.array([0.01, 0.5, 0.9999])
labels = ["0.01c", "0.5c", "0.9999c"]

gamma = 1 / np.sqrt(1 - beta**2)  #definig gamma
tau = gamma * tau0   # lifetime in our lab frame 

t = np.linspace(0, 200, 500) # timescale to check for variations change this to see blue line corresponding to 0.01c

plt.figure()
n1 = n0 * np.exp(-t / tau[0])
n2 = n0 * np.exp(-t / tau[1])  #seeing how N varies as we change beta in each case 
n3 = n0 * np.exp(-t / tau[2])
plt.plot(t, n1, label=labels[0])
plt.plot(t, n2, label=labels[1])
plt.plot(t, n3, label=labels[2])

plt.xlabel("Time in lab frame (units of Ï„0)")
plt.ylabel("N(t)/N0 (survival fraction)")
plt.title("Exponential decay for different velocities")
plt.legend()
plt.grid(True)
plt.show()
```
